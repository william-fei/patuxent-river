# Data Cleaning – Sora

```python
sora_full = pd.read_csv("Sora (Porzana carolina), Patuxent River Park, Maryland.csv")
```

```python
sora_full.columns
```

```python
print(str(sora_full['individual-local-identifier'].unique()))
```

Out of the 169 unique Sora ID, I manually went through the plots and decided to keep the following list of 34 sora for visualization:

```python
quality_list = ['1412-05692', '1412-08736', '1412-08744', '1412-08765', '1412-08788', '1412-08797', '1412-08842', '1412-08848', '1412-08858', '1412-08911', '1412-08912', '1412-58113', '1412-58120', '1412-58126', '1412-58169', '1412-58260', '1412-66717', '1412-66762', '1412-66778', '1412-66785', '1412-66788', '1412-80109', '1412-80111', '1412-80205', '1412-80208', '1412-80212', '1412-80213', '1412-80214', '1412-80215', '1412-80216', '1412-80218', '1412-80225', '1422-07028', '1422-07035']

filtered_sora = sora_full[sora_full['individual-local-identifier'].isin(quality_list)]
filtered_sora['individual-local-identifier'].unique()
```

```python
filtered_sora.head()
```

We will group by 'individual-local-identifier' then filter within each group by first unique location-long. This means dropping duplicates to only keep the first occurrence per bird ID.

```python
def filter_first_unique_location(group):
    return group.drop_duplicates(subset='location-long', keep='first')

filtered_sora_unique_locations = (
    filtered_sora
    .groupby('individual-local-identifier', group_keys=False)
    .apply(filter_first_unique_location)
)[['timestamp', 'location-long', 'location-lat', 'individual-local-identifier']]
```

```python
filtered_sora_unique_locations.head()
```

```python
filtered_sora_unique_locations['timestamp'] = pd.to_datetime(filtered_sora_unique_locations['timestamp'])
filtered_sora_unique_locations['timestamp'] = filtered_sora_unique_locations['timestamp'].apply(lambda x: x.replace(year=2022))
filtered_sora_unique_locations.head()
```

```python
sora_80214 = filtered_sora_unique_locations[filtered_sora_unique_locations['individual-local-identifier'] == '1412-80214']
```

# Memory Savings

```python
memory_usage_before_cleaning = sora_full.memory_usage(deep=True).sum() / (1024 ** 3)
memory_usage_after_cleaning = filtered_sora_unique_locations.memory_usage(deep=True).sum() / (1024 ** 3)
```

```python
print("Memory Usage Before Cleaning:", memory_usage_before_cleaning, "gb")
print("Memory Usage After Cleaning:", memory_usage_after_cleaning, "gb")
```

The total memory savings of this cleaning script was 99.9991563%.
